```python
import os
import shutil
from tqdm import tqdm
```

```python
source_dir = '/Users/nguyen/Documents/Work/Kaggle/Verum_raw_code'
des_dir = '/Users/nguyen/Documents/Work/Kaggle/Raw_code/data'
index = 1
batch_size = 1000
list_item = [i for i in os.listdir(source_dir) if '.c' in i]
len_item = len(list_item)
while len_item != 0:
    item_to_copy = list_item[(index - 1) * batch_size: index * batch_size]
    dir_ = f'batch_{index}'
    os.mkdir(os.path.join(des_dir + f'/{dir_}'))
    for item in tqdm(item_to_copy):
        source_file = source_dir + '/' + item
        des_file = des_dir + f'/{dir_}/' + item
        shutil.copy(source_file, des_file)
    len_item -= len(item_to_copy)
    index += 1
```

```python
total = 0
for d in os.listdir(des_dir):
    if 'batch' in d:
        print(len(os.listdir(os.path.join(des_dir, d))))
        total += len(os.listdir(os.path.join(des_dir, d)))
print(total)
```
